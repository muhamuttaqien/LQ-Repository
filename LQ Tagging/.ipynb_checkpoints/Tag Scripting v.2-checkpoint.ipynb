```python
import xlrd
import sqlite3
```

```python
# Open the workbook and define the worksheet

book = xlrd.open_workbook(filename="/Users/muhamuttaqien/Downloads/IndeksQuranTercanggih.xlsx")

grtopic = book.sheet_by_name("ILMU")
```

```python
conn = sqlite3.connect('/Users/muhamuttaqien/Downloads/IbnKathirTafsirs-formatted2 (1).db3')
c = conn.cursor()
```

### Count rows and columns

```python
table = "tafsir_id_content"
rowsQuery = "SELECT Count() FROM %s" % table
c.execute(rowsQuery)
numberOfRows = c.fetchone()[0]
numberOfRows
```

```python
columnsQuery = "PRAGMA table_info(%s)" % table
c.execute(columnsQuery)
numberOfColumns = len(c.fetchall())
numberOfColumns
```

```python
nrows = grtopic.nrows
nrows
```

```python
ncols = grtopic.ncols
ncols
```

### Clear all keywords (optional, hit only once)

```python
for row in range(numberOfRows+1):
    c.execute('UPDATE tafsir_id_content SET keywords = "" WHERE docid = ' + str(row))
    
conn.commit()
```

### 1. Detect parent and child topics

```python
topic = grtopic.cell_value(rowx=2, colx=1)
topic
```

```python
verse = grtopic.cell_value(rowx=8, colx=7)
verse
```

```python
# get verses
for pcol in range(7, ncols):
    colx = pcol
    verse = grtopic.cell_value(rowx=2, colx=colx).strip()
    if(verse):
        print(verse)
```

### 1st topics

```python
for prow in range(nrows):
    pcol = 1 # index where parent topics positioned
    topic = grtopic.cell_value(rowx=prow, colx=pcol)
    if(topic):
        print(topic.strip())
        srow = prow+1
        # get verses
        for pcol in range(7, ncols):
            colx = pcol
            verse = grtopic.cell_value(rowx=prow, colx=colx).strip()
            if(verse):
                # print(verse)
                # insert verse to db
                splitted_verse = verse[3:].split(":")
                surah = splitted_verse[0]
                ayah = splitted_verse[1]
                
                print("Tag into Qs. "+ surah + ":"+ ayah)                      
        
                # set multiple tagging
                for row in c.execute('SELECT keywords FROM tafsir_id_content WHERE c0sura = '+ surah +' AND c1aya = ' + ayah):
                    if(row[0] == None or row[0] == ''):
                        c.execute('UPDATE tafsir_id_content SET keywords = "'+ (topic + ' 1') +'" WHERE c0sura = '+ surah +' AND c1aya = ' + ayah)
                    else:
                        ct_topic = row[0] + ", " + (topic + ' 1')
                        print(ct_topic)
                        c.execute('UPDATE tafsir_id_content SET keywords = "'+ ct_topic +'" WHERE c0sura = '+ surah +' AND c1aya = ' + ayah)
                
conn.commit()
```
