### Check if empty

```python
def is_topic_empty(topic, depth_level, prow):
    
    if not topic:
        depth_level -=1
        pcol = depth_level
        topic = sheet.cell_value(rowx=prow, colx=pcol)
        
        topic, depth_level = is_topic_empty(topic, pcol, prow)

        return topic, depth_level
    else:
        return topic, depth_level
```

### Inject the verses

```python
def inject_the_verses(ncols, prow):
    verse_str = ""
    for pcol_v in range(7, ncols):
        colx = pcol_v
        verse = sheet.cell_value(rowx=prow, colx=colx).strip() # (Qs.2:38, Qs.2:62, Qs.2:112)
        if verse is not '':
            verse_str += " " + str(verse)
    
    return verse_str
```

### Tag topics to text

```python
def tag_topics(entry, sheet, depth_level=1):
    file = open("./topicstext/%s.txt" %entry, "w") 
    
    nrows = sheet.nrows
    ncols = sheet.ncols
    
    for prow in range(nrows):
        if(prow>1):
            pcol = depth_level
            topic = sheet.cell_value(rowx=prow, colx=pcol)
            
            if(topic[:3] == 'Qs.'):
                depth_level -=1
                pcol = depth_level
                topic = sheet.cell_value(rowx=prow, colx=pcol)
                
            verse_str = inject_the_verses(ncols, prow) # it returns [Qs.2:38, Qs.2:62, Qs.2:112]
            
            topic = re.sub(r"^\s+", "", topic, flags=re.UNICODE)

            if(topic):             
                topic = re.sub(r"^\s+", "", topic, flags=re.UNICODE)
                topic = str(depth_level) + " " + topic + verse_str + "\n"
                print(topic)
                file.write(topic)
                
                depth_level +=1
            else:
                depth_level -=1
                pcol = depth_level
                topic = sheet.cell_value(rowx=prow, colx=pcol)
                
                topic, depth_level = is_topic_empty(topic, pcol, prow)
                        
                topic = re.sub(r"^\s+", "", topic, flags=re.UNICODE)
                topic = str(depth_level) + " " + topic + verse_str + "\n"
                print(topic)
                file.write(topic)

                depth_level +=1
    file.close()
```

```python
entry = "IMAN"
sheet = BOOK.sheet_by_name(entry)
tag_topics(entry, sheet, 1)
```

```python
LIST_ENTRY = ["IMAN", "ILMU", "BANGSA-BANGSA TERDAHULU", "SEJARAH", "AL QUR'AN", 
           "AKHLAQ & ADAB", "IBADAH", "MAKANAN & MINUMAN", "PAKAIAN & PERHIASAN", 
           "HUKUM PRIVAT", "MUAMALAT", "PERADILAN & HAKIM", "HUKUM PIDANA, JINAYAH", "JIHAD"]

for entry in LIST_ENTRY:
    sheet = BOOK.sheet_by_name(entry)
    tag_topics(entry, sheet, 1)
```
