#### maududi

```python
f = open(os.path.join('./maududi-en.txt'),"r")
```

```python
for (i, line) in enumerate(f):
    text = line.split('|')
    text_no_apostrophe = text[2].replace("'", "\\'")
    query = "UPDATE translation_en SET ala_maududi ='" + text_no_apostrophe + "' WHERE chapter_id = " + text[0] + " AND verse_id = " + text[1]

    print("id %d is updated..." %i)
    
    c.execute(query)
```

```python
db.commit()
```
